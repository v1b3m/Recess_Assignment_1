```python
import pandas as pd
```

```python
# import the dataset
train = pd.read_csv('Dataset/Additional_Country_level_data.csv')
```

```python
train.head()
```

```python
# Remove duplicate rows from the DataFrame
train = train.drop_duplicates()
```

```python
# delete the government effectiveness index in 2005 column since all its values are undefined
train.drop('government effectiveness index in 2005', axis=1, inplace=True)
```

```python
# Confirm that the column has been removed
train.head()
```

```python
# Get columns with string data
categorical_cols = [cname for cname in train.columns if train[cname].dtype == "object"]
```

```python
categorical_cols
```

```python
# Perfom label encoding on categorical data
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_train = train.copy()
label_train['country'] = label_encoder.fit_transform(train['country'])
```

```python
label_train.head()
```

```python
# Perform feature scaling on the data
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
scaled_train = ss.fit_transform(label_train)
```
